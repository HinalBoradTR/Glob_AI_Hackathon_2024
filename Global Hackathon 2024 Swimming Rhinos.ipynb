```python
import snowflake.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
```

```python

# Connect to Snowflake
conn = snowflake.connector.connect(
    user='hinal.borad@thomsonreuters.com',
    authenticator='externalbrowser',
    account='A206448_PROD.us-east-1',
    warehouse='A208192_GLOBALAIHACKATHON_DATAANALYTICS_MDS_WH',
    database='MYDATASPACE',
    schema='A208192_GLOBALAIHACKATHON_DATAANALYTICS'
)

# Function to execute query and return a DataFrame
def execute_query(query):
    cur = conn.cursor()
    cur.execute(query)
    df = cur.fetch_pandas_all()
    cur.close()
    return df

# Example query (customize as needed)
query = "SELECT * FROM FEATURE_EVENT LIMIT 1000;"
df_feature_event = execute_query(query)

# Display basic data statistics
print(df_feature_event.describe())

# Plot example: Number of events over time
plt.figure(figsize=(10, 6))
df_feature_event['TIMESTAMP'] = pd.to_datetime(df_feature_event['TIMESTAMP'])
df_feature_event.set_index('TIMESTAMP')['NUM_EVENTS'].plot()
plt.title('Number of Events Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Number of Events')
plt.show()
```

```python
# AI Solution 1: Anomaly Detection
iso_forest = IsolationForest(n_estimators=100)
df_feature_event['anomaly'] = iso_forest.fit_predict(df_feature_event[['NUM_EVENTS']])
df_feature_event['anomaly'] = df_feature_event['anomaly'].map({1: 0, -1: 1})  # Convert -1/1 to 0/1 for clarity

# Plot anomalies in the data
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_feature_event.index, y=df_feature_event['NUM_EVENTS'], hue=df_feature_event['anomaly'], palette=['green', 'red'])
plt.title('Anomaly Detection in Number of Events')
plt.xlabel('Timestamp')
plt.ylabel('Number of Events')
plt.show()

# AI Solution 2: Simple NLP Task - Extracting Keywords from Poll Questions
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

# Fetching poll data
query_polls = "SELECT QUESTION FROM POLLS LIMIT 1000;"
df_polls = execute_query(query_polls)

# Vectorize questions to find common words
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(df_polls['QUESTION'])
words = vectorizer.get_feature_names_out()
word_counts = X.toarray().sum(axis=0)
word_freq = dict(zip(words, word_counts))

# Display top 10 common words in questions
top_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)[:10]
print("Top 10 common words in poll questions:")
for word, freq in top_words:
    print(f"{word}: {freq}")

```

```python
sns.pairplot(df_feature_event, hue='anomaly', palette='bright')
plt.suptitle('Pair Plot of Features', y=1.02)
plt.show()
```

```python
# Filter only numeric columns for correlation matrix
numeric_df = df_feature_event.select_dtypes(include=[np.number])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#### 3. Histogram with KDE
#### Visualize the distribution of a particular numerical variable.
plt.figure(figsize=(10, 6))
sns.histplot(df_feature_event['NUM_EVENTS'], kde=True, color='magenta')
plt.title('Distribution of Number of Events with KDE')
plt.xlabel('Number of Events')
plt.ylabel('Frequency')
plt.show()
```
